```python
import pandas as pd
```

```python
d = pd.read_csv('market.csv')
```

```python
data = d.iloc[:, :8]
```

```python
data.head(3)
```

```python
# Set 'EET' as the index for the time periods
data.set_index('EET', inplace=True)

# Transpose the data to switch columns and rows (dates will become columns)
data = data.T

# Rename the columns to make them easier to work with
data.columns = [f"Hour_{i+1}" for i in range(data.shape[1])]

# Reset index to convert the dates into a regular column
data.reset_index(inplace=True)

# Rename the index column to 'Date'
data.rename(columns={'index': 'Date'}, inplace=True)

# Display cleaned data
data.head()
```

```python
data = data.iloc[:7, :25]
data.head()
```

```python
import matplotlib.pyplot as plt

# Plot price trends for each day
plt.figure(figsize=(10, 6))
for i, row in data.iterrows():
    plt.plot(range(24), row[1:], label=row['Date'])

plt.xlabel('Hour of the Day')
plt.ylabel('Price (£)')
plt.title('Electricity Prices Over Time')
plt.legend(loc='upper right')
plt.show()
```

```python
# Check the data types and look for any formatting issues
data.info()

# Also, check a few rows for any special characters or extra spaces
data.head()
```

```python
# Remove any non-numeric characters (like commas, currency symbols, etc.) from each column
for col in data.columns[1:]:
    # Remove commas, currency symbols, and spaces
    data[col] = data[col].replace({'£': '', ',': '', ' ': ''}, regex=True)

# Convert columns to numeric
for col in data.columns[1:]:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Check if the conversion worked
data.head()
```

```python
# Calculate percentage change for each hour's price
data_pct_change = data.set_index('Date').pct_change(axis=1)

# Calculate volatility (standard deviation) for each day
volatility = data_pct_change.std(axis=1)

# Display volatility
print(volatility)
```

***analyzing volatility***

```python
# Sort days by volatility to identify the most and least volatile
volatility_sorted = volatility.sort_values(ascending=False)

# Display the top 3 most volatile days
print("Top 3 most volatile days:")
print(volatility_sorted.head(3))

# Display the top 3 least volatile days
print("\nTop 3 least volatile days:")
print(volatility_sorted.tail(3))
```

```python
# Extract the data for the top 3 most volatile days
top_volatile_days = data[data['Date'].isin(volatility_sorted.head(3).index)]

# Plot the price changes for the most volatile days
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
for i, row in top_volatile_days.iterrows():
    plt.plot(range(24), row[1:], label=row['Date'])

plt.xlabel('Hour of the Day')
plt.ylabel('Price (£)')
plt.title('Price Changes for the Top 3 Most Volatile Days')
plt.legend(loc='upper right')
plt.show()
```

```python
# Define thresholds for buying and selling
buy_threshold = -0.02  # Buy when the price drops by 2% or more
sell_threshold = 0.02  # Sell when the price rises by 2% or more

# Simulate a trading strategy based on percentage changes
trades = []
for i in range(len(data_pct_change)):  # Use range(len()) to ensure integer-based indexing
    for j in range(1, len(data_pct_change.columns)):  # Iterate over columns
        # Get the percentage change at [i, j]
        pct_change = data_pct_change.iloc[i, j]
        
        # Get the corresponding price from the original data
        price = data.iloc[i, j+1]  # Use iloc with integer indexing
        
        # Check the buy or sell condition
        if pct_change <= buy_threshold:  # Buy action
            trades.append({'Date': data_pct_change.index[i], 'Hour': j, 'Action': 'Buy', 'Price': price})
        elif pct_change >= sell_threshold:  # Sell action
            trades.append({'Date': data_pct_change.index[i], 'Hour': j, 'Action': 'Sell', 'Price': price})

# Convert the trades list to a DataFrame for better analysis
trades_df = pd.DataFrame(trades)

# Display the trade actions
trades_df.head()
```

```python
# Estimate profit by pairing buy and sell actions (simplified)
buy_prices = trades_df[trades_df['Action'] == 'Buy']['Price'].reset_index(drop=True)
sell_prices = trades_df[trades_df['Action'] == 'Sell']['Price'].reset_index(drop=True)

# Calculate profits (assuming each buy is followed by a sell)
profits = sell_prices - buy_prices

# Total profit for the week
total_profit = profits.sum()
print(f"Estimated total profit: £{total_profit}")
```
